```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

# Read in the output
with open('/home/kevin/Documents/udacity_car/CarND-Unscented-Kalman-Filter-Project-master/build/output.txt') as f:
    df = pd.read_table(f, sep='\t', header=0, lineterminator='\n')
```

```python
# Check to make sure the read was correct
print(df.head())
```

```python
# Save chi squared expectations for each sensor type (95% should be lower than this threshold)
e_radar = 7.815
e_laser = 5.991

# Plot radar NIS
r = df[df["sensor"] == "radar"]["NIS"]
r.plot(title="Radar")
plt.plot((0, r.index.max()), (e_radar, e_radar), "red")
plt.show()
plt.clf()
print("{}% of values are greater than the 95.00% threshold.".format(round(len(r[r>e_radar].index)/len(r.index)*100, 2)))

# Plot lidar NIS
r = df[df["sensor"] == "laser"]["NIS"]
r.plot(title="Laser")
plt.plot((0, r.index.max()), (e_laser, e_laser), "red")
plt.show()
plt.clf()
print("{}% of values are greater than the 95.00% threshold.".format(round(len(r[r>e_laser].index)/len(r.index)*100, 2)))
```
